```python
new_df = new_df.dropna()
print(new_df.shape)
print(new_df.isna().sum())
```

```python
# Save the updated dataset
new_df.to_csv('updated_beats_review_dataset.csv', index=False)
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from textblob import TextBlob
```

```python
# Create new columns for polarity and subjectivity
new_df['Polarity'] = new_df['content'].apply(lambda x: TextBlob(x).sentiment.polarity)
new_df['Subjectivity'] = new_df['content'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Preview the updated dataset
new_df.head()
```

```python
def categorize_sentiment(polarity):
    if polarity > 0.2:
        return 'Positive'
    elif polarity < -0.2:
        return 'Negative'
    else:
        return 'Neutral'

# Apply categorization
new_df['Sentiment'] = new_df['Polarity'].apply(categorize_sentiment)

# Count the number of each sentiment
print(new_df['Sentiment'].value_counts())
```

```python
new_df.to_csv('updated_beats_review_dataset.csv', index=False)
```

```python
import re

# Extract fields using regular expressions
def extract_attr(text, key):
    pattern = rf"{key}:\s*([^:]+?)(?=([A-Z][a-z]+:|$))"
    match = re.search(pattern, text)
    return match.group(1).strip() if match else None

new_df["Color"] = new_df["product_attributes"].apply(lambda x: extract_attr(x, "Color"))
new_df["Style"] = new_df["product_attributes"].apply(lambda x: extract_attr(x, "Style"))
new_df["Set"] = new_df["product_attributes"].apply(lambda x: extract_attr(x, "Set"))
```

```python
new_df.head()
```

```python
sent_map = {"Positive": 1, "Neutral": 0, "Negative": -1}
new_df["SentimentScore"] = new_df["Sentiment"].map(sent_map)
new_df.head()
```

```python
new_df["ReviewLength"] = new_df["content"].str.len()
new_df.head()
```

```python
def star_bucket(r):
    if r <= 2:
        return "Low (1–2)"
    elif r == 3:
        return "Neutral (3)"
    else:
        return "High (4–5)"

new_df["RatingBucket"] = new_df["rating"].apply(star_bucket)
new_df.head()
```

```python
new_df["mentions_fit"] = new_df["content"].str.contains("fit|fitting|ear tip|fall", case=False, regex=True)
new_df["mentions_anc"] = new_df["content"].str.contains("ANC|noise|cancel", case=False, regex=True)
new_df["mentions_battery"] = new_df["content"].str.contains("battery|charge", case=False, regex=True)
new_df["mentions_mic"] = new_df["content"].str.contains("mic|microphone|call", case=False, regex=True)
new_df["mentions_connect"] = new_df["content"].str.contains("connect|bluetooth", case=False, regex=True)
new_df.head()
```

```python
new_df.to_csv('beats_reviews_for_tableau.csv', index=False)
```

```python
topic_cols = [
    "mentions_fit",
    "mentions_anc",
    "mentions_battery",
    "mentions_mic",
    "mentions_connect"
]
```
